## Display the database

```python
import json
import numpy as np
import pandas as pd

with open('materials.json', 'rb') as f:
    data = json.load(f)
```

## Phonons

```python
X = []
Name = []
for i in range(0,len(data)):
    freq = []
    for j in ['Gamma','X','M','R']:
        freq = freq + data[i]['phonon'][j]['freq']
    X.append(freq)
    Name.append(str(data[i]['name']))

X = np.array(X)
Name = np.array([Name])

Phonons = np.concatenate((Name.T, X), axis=1)
pd.DataFrame(Phonons)
```

### Born Effective Charges

```python
X = []
Name = []
for i in range(10):
    Name.append(str(data[i]['name']))
    bec = []
    for j in ['0','1','2','3','4']:
        bec = bec + data[i]['becs'][j]
    X.append(bec)

X = np.reshape(np.array(X), (10, -1))
Name = np.array([Name])

becs = np.concatenate((Name.T, X), axis=1)
pd.DataFrame(becs)
```
